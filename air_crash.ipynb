```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv("train.csv")
df.head()
```

```python
df.corr()
```

# Separate input and target data

```python
y=df['Severity']
del df['Severity']
del df['Accident_ID']
del df['Accident_Type_Code']
```

# Data Normalisation

```python
for col in df.columns:
    df[col]=df[col]/df[col].mean()
df.head()
    
```

# Splitting Data into Training and Testing 

```python
X_train, X_test, y_train, y_test = train_test_split(
    df, y, test_size=0.33, random_state=42)
```

# Training and testing on KNN

```python
from sklearn.neighbors import KNeighborsClassifier
max_accuracy=0
neighbour=0
for n in range(1,50):
    neigh = KNeighborsClassifier(n_neighbors=n)
    neigh.fit(X_train, y_train)
    y_pred=neigh.predict(X_test)
    accuracy=accuracy_score(y_test, y_pred)
    if accuracy>max_accuracy:
        max_accuracy=accuracy
        neighbour=n
print ("Maximum accuracy:",max_accuracy,' on Neighbours=',neighbour)
```

# Training and Testing on Decision Tree

```python

clf = tree.DecisionTreeClassifier()

clf = clf.fit(X_train, y_train)

y_pred=clf.predict(X_test)

print ("Accuracy on Decision Tree:", accuracy_score(y_test, y_pred))
```

# Training and Testing on Random Forest

```python
clf = RandomForestClassifier(max_depth=29, random_state=9)
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
accuracy=accuracy_score(y_test, y_pred)
print ("Accuracy Score on Random Forest Classifier:",accuracy)
```
